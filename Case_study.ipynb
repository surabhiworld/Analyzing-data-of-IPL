```python
#Loadng the equired libraries

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading the data set
ipl = pd.read_csv('https://raw.githubusercontent.com/surabhiworld/Analyzing-data-of-IPL/main/IPL_data.csv')
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
#Getting the frequency of most man of the matches award
ipl['player_of_match'].value_counts()
```

We can indentify that CH Gayle has won 17 man of the matches and so on.

```python
# Want to see only top 5 records.
ipl['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize = (8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color = 'purple')
plt.show
```

On the basis og graph we can see that CH Gayle has won most number of matches awards

```python
#Getting the frequency of result column
ipl['result'].value_counts()
```

There are three result:
1.Normal
2.Tie
3.no result

568 matches were normal in total matches

```python
#Getting the frequency of toss winner
ipl['toss_winner'].value_counts()
```

So in toss winner we can see that Mumbai indian won more number of tosses as compare to others

```python
#Extracting the records where a team won betting first
batting_first = ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize = (5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Ditribution of runs')
plt.xlabel('Runs')
plt.show
```

Per the graph we can indetify that more that 100 team won the matches by arround 20 runs and arround 10 team won the matches by 140 runs.




```python
# gettng frequecy of winner in betting_first
batting_first['winner'].value_counts()
```

```python
#making the bar plot for top 3 teams which most wins after batting first
plt.figure(figsize=(5,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color = 'b')
plt.show
```

```python
# Extracting those records where a team has won after batting second
batting_second = ipl[ipl['win_by_wickets']!=0]
batting_second
```

```python
#finding the frequency of number of winner after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top 5 teams
plt.figure(figsize=(12,10))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]),color = 'g')
plt.show
```
